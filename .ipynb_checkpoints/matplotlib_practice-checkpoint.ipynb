```python
%matplotlib inline
```

```python
# Import dependenies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.84, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() --- 'ax.plot()' is the object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Another way to create a line chart using object-oriented interface method
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() using object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# to change figure-level attributes (e.g., axis labels, a title, or a legend) or save the figure as an image
fig = plt.figure()
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot using object-oriented method
ax = plt.axes()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')
ax.set_ylim(0, 45)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a bar chart using the object-oriented method
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# If we want the data on opposite axes, we need to switch the arguments
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend()
plt.gca().invert_yaxis()
```

```python
# There are 2 ways to create scatter plots using MATLAB method
# plt.plot()
# plt.scatter()
# Create a scatter plot using the MATLAB method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
# When using plt.plot() function to create scatter plot we need to add a lowercase "o" inside the parenthesis.
# This switches from a line chart to a scatter plot.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, 'o', color="red", label='Chicago')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0, 50)
plt.legend()
```

```python
# Create a Bubble Chart using MATLAB Method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refactor code to replace for loop
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])
```

```python
# Create a scatter plot using Object-Oriented interface
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create Bubble chart using Oject_Oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create Bubble chart using Oject_Oriented interface
fig, ax = plt.subplots()
ax.scatter(y_axis, 
           x_axis, 
           color="skyblue"  
           alpha=0.2,
           edgecolor="black", 
           linewidth=2, 
           s = [i * 5 for i in y_axis], 
           label="Chicago")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Create pie chart using MATLAB method.
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
plt.show()
```

```python
# Create pie chart using Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", 
          "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       shadow=True,
       startangle=90,
       counterclock=False,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```
